```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Meta-Llama-3.1-8B",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # use one if using gated models like meta-llama/Llama-2-7b-hf
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # Supports any, but = 0 is optimized
    bias = "none",    # Supports any, but = "none" is optimized
    # [NEW] "unsloth" uses 30% less VRAM, fits 2x larger batch sizes!
    use_gradient_checkpointing = "unsloth", # True or "unsloth" for very long context
    random_state = 3407,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
)
```

```python
alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""

# Formatting the dataset
def formatting_prompts_func(examples):
    instructions = examples["title"]
    inputs       = examples["genre"]
    outputs      = examples["story"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        # Create the prompt with the alpaca format
        text = alpaca_prompt.format(instruction, input, output)
        texts.append(text)
    return {"text": texts}

# Loading the dataset and formatting the text
from datasets import load_dataset
ds = load_dataset("FareedKhan/1k_stories_100_genre", split="train")
dataset = ds.map(formatting_prompts_func, batched=True)

# Tokenizing the dataset and adding 'labels'
def tokenize_function(examples):
    tokenized_output = tokenizer(examples["text"], padding="max_length", truncation=True, max_length=512)

    # For language modeling, set the 'labels' to be the same as 'input_ids'
    tokenized_output["labels"] = tokenized_output["input_ids"].copy()  # Copy the input_ids to labels

    return tokenized_output

# Tokenizing the dataset
tokenized_dataset = dataset.map(tokenize_function, batched=True)

from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments, Trainer
training_args = TrainingArguments(
    output_dir="./results",               # Directory to save checkpoints
    evaluation_strategy="no",             # No evaluation during training since there's no test set
    learning_rate=5e-5,                   # Learning rate
    per_device_train_batch_size=8,        # Batch size for training
                   # Number of epochs to train
    max_steps=170,
    weight_decay=0.01,                    # Weight decay for regularization
    logging_dir="./logs",                 # Directory for logs
    save_total_limit=2,                   # Limit the number of checkpoints
    fp16=True                             # Use mixed precision training if available
)

# Initialize Trainer (without eval_dataset)
trainer = Trainer(
    model=model,                          # The model to train
    args=training_args,                   # Training arguments defined above
    train_dataset=tokenized_dataset,  # Training dataset (entire dataset)
    #tokenizer=tokenizer                   # Tokenizer for input data
)


```

```python
# Now proceed with training
trainer.train()
```

```python
from unsloth.models import FastLanguageModel
model = FastLanguageModel.for_inference(model)


input_text=" a horror  story with 2 caharcters "
inputs = tokenizer(input_text, return_tensors="pt")
output = model.generate(**inputs, max_length=200, temperature=0.7, top_k=50, top_p=0.9)
output_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(output_text)

```
